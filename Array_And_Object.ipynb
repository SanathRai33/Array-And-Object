### **Q1. In the following shopping cart add, remove, and edit items**
   => const shoppingCart = ['Milk', 'Coffee', 'Tea', 'Honey']
- add 'Meat' in the beginning of your shopping cart if it has not been already added.
- add Sugar at the end of you shopping cart if it has not been already added/
- remove 'Honey' if you are allergic to honey.
- modify Tea to 'Green Tea'.

<br/>

**Answer-->**

**Shopping cart operations:**
```
let shoppingCart = ['Milk', 'Coffee', 'Tea', 'Honey'];
```

**1. Add 'Meat' at the beginning if not present:**
```
if (!shoppingCart.includes('Meat')) {
  shoppingCart.unshift('Meat');
}
```

**2. Add 'Sugar' at the end if not present:**
```
if (!shoppingCart.includes('Sugar')) {
  shoppingCart.push('Sugar');
}
```

**3. Remove 'Honey' if allergic:**
```
let allergicToHoney = true;
if (allergicToHoney) {
  shoppingCart = shoppingCart.filter(item => item !== 'Honey');
}
```

**4. Modify 'Tea' to 'Green Tea':**
```
const teaIndex = shoppingCart.indexOf('Tea');
if (teaIndex !== -1) {
  shoppingCart[teaIndex] = 'Green Tea';
}
console.log("Updated Shopping Cart:", shoppingCart);
```

### **Q2. The following is an array of 10 students ages:**
 => const ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24O]
- Sort the array and find the min and max age.
- Find the median age(one middle item or two middle items divided by two)
- Find the average age(all items divided by number of items)
- Find the range of the ages(max minus min)
- Compare the value of (min - average) and (max - average), use abs() method

<br/>

**Answer-->**

```
const ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24];
```

**1. Sort:**
```
const sortedAges = [...ages].sort((a, b) => a - b);
console.log("Sorted Ages:", sortedAges);
```

**2. Min and Max:**
```
const minAge = sortedAges[0];
const maxAge = sortedAges[sortedAges.length - 1];
console.log("Min Age:", minAge, "Max Age:", maxAge);
```

**3. Median:**
```
let median;
const mid = Math.floor(sortedAges.length / 2);
if (sortedAges.length % 2 === 0) {
  median = (sortedAges[mid - 1] + sortedAges[mid]) / 2;
} else {
  median = sortedAges[mid];
}
console.log("Median Age:", median);
```

**4. Average:**
```
const average = ages.reduce((acc, val) => acc + val, 0) / ages.length;
console.log("Average Age:", average.toFixed(2));
```

**5. Range:**
```
const range = maxAge - minAge;
console.log("Age Range:", range);
```

**6. Compare abs(min - average) and abs(max - average):**
```
console.log("Min - Avg:", Math.abs(minAge - average).toFixed(2));
console.log("Max - Avg:", Math.abs(maxAge - average).toFixed(2));
```

### **Q3. Object Extensibility and Sealing**
a) Use the Object.preventExtensions method to prevent any further additions of properties to the student object.

b) Use the Object.isExtensible method to check if the student object is extensible. Store the result in a variable called extensibleStatus.

c) Create a new object called teacher with a 'subject' property set to 'Math'.

d) Use the Object.seal method to seal the teacher object, preventing any additions or deletions of properties.

e) Use the Object.isSealed method to check if the teacher object is sealed. Store the result in a variable called sealedStatus.

f) Print the extensibleStatus and sealedStatus to the console.

<br/>

**Answer-->**

 **a) Prevent extensions:**
 ```
const student = { name: "Alex", age: 21 };
Object.preventExtensions(student);
```

 **b) Check extensibility:**
 ```
const extensibleStatus = Object.isExtensible(student);
```

 ```
const teacher = { subject: "Math" };
```

 ```
const sealedStatus = Object.isSealed(teacher);
```

### **Q4. Assignment: Building a Student Management System**

**Description:**<br/>
You are tasked with building a student management system using JavaScript. The system should allow you to perform various operations on a list of students, including adding, updating, deleting, and displaying student information.

<br/>

**Requirements:**<br/>
Here is an initial array of students. Each student is represented as an object with the following properties: id,
firstName, lastName, age, and grade.

```
const students = [

  { id: 1, firstName: "John", lastName: "Doe", age: 20, grade: "A" },
  { id: 2, firstName: "Jane", lastName: "Smith", age: 22, grade: "B" },
  { id: 3, firstName: "Bob", lastName: "Johnson", age: 19, grade: "A" },

  // Add more students as needed
];
```

<br/>

**Answer-->**

```
let students = [
  { id: 1, firstName: "John", lastName: "Doe", age: 20, grade: "A" },
  { id: 2, firstName: "Jane", lastName: "Smith", age: 22, grade: "B" },
  { id: 3, firstName: "Bob", lastName: "Johnson", age: 19, grade: "A" },
];

let student = { id: 4, firstName: "Alice", lastName: "Wonder", age: 21, grade: "B" };
```

**a) Add Student:**
```
function addStudent(student) {
  students.push(student);
}
```

**b) Update Student:**
```
function updateStudent(id, newData) {
  const index = students.findIndex(s => s.id === id);
  if (index !== -1) {
    students[index] = { ...students[index], ...newData };
  }
}
```

**c) Delete Student:**
```
function deleteStudent(id) {
  students = students.filter(s => s.id !== id);
}
```

**d) List Students:**
```
function listStudents() {
  console.log("All Students:");
  students.forEach(s => console.log(s));
}
```

**e) Find by Grade:**
```
function findByGrade(grade) {
  return students.filter(s => s.grade === grade);
}
```

**f) Calculate Average Age:**
```
function averageAge() {
  const total = students.reduce((sum, s) => sum + s.age, 0);
  return (total / students.length).toFixed(2);
}
```

**Test Cases:**
```
addStudent({ id: 4, firstName: "Alice", lastName: "Wonder", age: 21, grade: "B" });

updateStudent(2, { age: 23 });

deleteStudent(1);

listStudents();

console.log("Grade A Students:", findByGrade("A"));

console.log("Average Age:", averageAge());
```
